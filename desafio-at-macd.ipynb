```python
import pandas as pd
import numpy as np
import mplfinance as mpf
import ta
import yfinance as yf
```

```python
df = yf.download('BTC-USD', start='2022-01-01')
```

```python
df['mm_fast'] = df['Close'].rolling(12).mean()
df['mm_slow'] = df['Close'].rolling(26).mean()
```

```python
df['macd'] = ta.trend.MACD(df['Close']).macd()
df['macd_signal'] = ta.trend.MACD(df['Close']).macd_signal()
```

```python
df_plot = df.loc['2023-02-01':]
plot_mm_fast = mpf.make_addplot(df_plot['mm_fast'])
plot_mm_slow = mpf.make_addplot(df_plot['mm_slow'])
plot_macd = mpf.make_addplot(df_plot['macd'], panel=1, color='red')
plot_macd_signal = mpf.make_addplot(df_plot['macd_signal'], panel=1)

mpf.plot(df_plot, type='candle', addplot=[plot_mm_fast, plot_mm_slow, plot_macd, plot_macd_signal], figsize=(16,9))
```
